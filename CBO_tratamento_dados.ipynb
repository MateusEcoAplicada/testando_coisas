```python
import re

def descompactar_cargos(df):
    """
    Descompacta ocupações que estão em uma única linha, separadas por \n
    Mantém alinhamento entre título e descrição
    """
    rows = []
    
    for idx, row in df.iterrows():
        titulo = str(row['titulo']).strip()
        descricao = str(row['descricao']).strip()
        
        # Divide títulos por quebra de linha
        titulos_lista = [t.strip() for t in titulo.split('\n') if t.strip()]
        
        # Se houver múltiplos títulos, divide a descrição também (ou repete)
        if len(titulos_lista) > 1:
            # Tenta dividir a descrição por padrão de código ou usa a mesma para todos
            descricoes_lista = descricao.split('\n') if '\n' in descricao else [descricao] * len(titulos_lista)
            
            # Se números de itens não baterem, repete a última descrição
            while len(descricoes_lista) < len(titulos_lista):
                descricoes_lista.append(descricoes_lista[-1] if descricoes_lista else descricao)
        else:
            titulos_lista = [titulo]
            descricoes_lista = [descricao]
        
        # Cria uma linha para cada título
        for t, d in zip(titulos_lista, descricoes_lista):
            rows.append({'titulo': t.strip(), 'descricao': d.strip()})
    
    return pd.DataFrame(rows)

# Aplica a descompactação
df_descompactado = descompactar_cargos(df)

df_descompactado.info()
```

```python
def descompactar_descricoes_correto(df):
    """
    Descompacta títulos que estão em uma única linha com \n
    Separa ocupações pelo padrão NNNN-NN (CBO code)
    Mantém a mesma descrição para todas as ocupações do mesmo grupo
    Remove quebras de linha da descrição (normaliza espaços)
    """
    import re
    rows = []
    
    for idx, row in df.iterrows():
        titulo = str(row['titulo']).strip()
        descricao = str(row['descricao']).strip()
        
        # Se descricao for 'nan', deixa vazia
        if descricao.lower() == 'nan':
            descricao = ''
        else:
            # Remove quebras de linha da descrição e normaliza espaços
            descricao = ' '.join(descricao.split())
        
        # Separa ocupações pelo padrão NNNN-NN (CBO code no início)
        # Encontra todos os matches de "NNNN-NN" seguidos de tudo até o próximo NNNN-NN ou fim
        ocupacoes = []
        
        # Primeiro, substitui \n por espaço para normalizar
        titulo_normalizado = ' '.join(titulo.split('\n'))
        
        # Padrão: NNNN-NN no início seguido de resto até o próximo NNNN-NN
        padrao = r'(\d{4}-\d{2}[^\d]*?)(?=\d{4}-\d{2}|$)'
        matches = re.findall(padrao, titulo_normalizado)
        
        if matches:
            ocupacoes = [m.strip() for m in matches if m.strip()]
        else:
            # Se não encontrar padrão, usa a string inteira
            ocupacoes = [titulo]
        
        # Cria uma linha para cada ocupação com a mesma descrição
        for ocupacao in ocupacoes:
            rows.append({
                'titulo': ocupacao,
                'descricao': descricao
            })
    
    return pd.DataFrame(rows)

# Aplica a descompactação
df_descompactado = descompactar_descricoes_correto(df)

df_descompactado.info()
```

```python
# Descompactar linhas com padrão NNNNNN (sem hífen)
def descompactar_sem_hifen(texto, descricao):
    """
    Descompacta ocupações com padrão NNNNNN (6 dígitos sem hífen)
    """
    import re
    
    # Normaliza quebras de linha em espaços
    texto_normalizado = ' '.join(texto.split('\n'))
    
    # Padrão: 6 dígitos seguidos de resto até próximos 6 dígitos ou fim
    padrao = r'(\d{6}[^\d]*?)(?=\d{6}|$)'
    matches = re.findall(padrao, texto_normalizado)
    
    if matches:
        ocupacoes = [m.strip() for m in matches if m.strip()]
    else:
        ocupacoes = [texto]
    
    # Normaliza descrição
    descricao_normalizada = ' '.join(str(descricao).split())
    
    return [{
        'titulo': ocp,
        'descricao': descricao_normalizada
    } for ocp in ocupacoes]

# Pegar as linhas com NaN e descompactar
linhas_nan = dados[dados['ocupacao'].isna()].copy()

linhas_descompactadas = []
for idx, row in linhas_nan.iterrows():
    linhas_descompactadas.extend(
        descompactar_sem_hifen(row['titulo'], row['descricao'])
    )

# Criar DataFrame com linhas descompactadas
df_novas = pd.DataFrame(linhas_descompactadas)

# Adicionar coluna 'ocupacao' para as novas linhas (formato NNNNNN sem hífen)
df_novas['ocupacao'] = df_novas['titulo'].str.extract(r'(\d{6})')[0]

print(f"Linhas descompactadas: {len(df_novas)}")
print("\nNovas linhas:")
df_novas
```

```python
# Criar coluna descricao em lista_não_tem_df com base no Grupo (4 primeiros dígitos)
lista_não_tem_df['grupo_4d'] = lista_não_tem_df['ocupacao'].astype(str).str[:4]

# Mapeamento Grupo -> Tarefa (usado como descrição)
# Quando há mais de uma tarefa no mesmo grupo, usa a mais frequente
descricao_por_grupo = (
    dados_nossos.groupby('Grupo')['Tarefa']
    .agg(lambda s: s.mode().iat[0] if not s.mode().empty else s.dropna().iat[0] if not s.dropna().empty else '')
)

lista_não_tem_df['descricao'] = lista_não_tem_df['grupo_4d'].map(descricao_por_grupo).fillna('')

# (opcional) remover coluna auxiliar
lista_não_tem_df.drop(columns=['grupo_4d'], inplace=True)

print('Coluna descricao criada com sucesso.')
print(lista_não_tem_df[['ocupacao', 'titulo', 'tarefa', 'descricao']].head(15))
print('\nDescrições vazias:', (lista_não_tem_df['descricao'] == '').sum())
```

```python
# Preencher coluna tarefa em `dados` usando `dados_nossos`
# Regra: 1) match exato ocupacao == CODIGO; 2) fallback por grupo (4 primeiros dígitos)

dados_nossos_aux = dados_nossos.copy()
dados_nossos_aux['CODIGO'] = dados_nossos_aux['CODIGO'].astype(str).str.zfill(6)
dados_nossos_aux['Grupo'] = dados_nossos_aux['Grupo'].astype(str).str.zfill(4)

dados['ocupacao'] = dados['ocupacao'].astype(str).str.zfill(6)

# 1) Match exato por código
tarefa_por_codigo = (
    dados_nossos_aux
    .dropna(subset=['Tarefa'])
    .drop_duplicates(subset=['CODIGO'])
    .set_index('CODIGO')['Tarefa']
)

dados['tarefa'] = dados['ocupacao'].map(tarefa_por_codigo)

# 2) Fallback por grupo (4 dígitos)
tarefa_por_grupo = (
    dados_nossos_aux
    .dropna(subset=['Tarefa'])
    .groupby('Grupo')['Tarefa']
    .agg(lambda s: s.mode().iat[0] if not s.mode().empty else s.iloc[0])
)

mask_faltando = dados['tarefa'].isna()
dados.loc[mask_faltando, 'tarefa'] = (
    dados.loc[mask_faltando, 'ocupacao'].str[:4].map(tarefa_por_grupo)
)

print('Coluna tarefa preenchida em `dados`.')
print('Total linhas:', len(dados))
print('Tarefas preenchidas:', dados['tarefa'].notna().sum())
print('Tarefas vazias:', dados['tarefa'].isna().sum())
dados[['ocupacao', 'titulo', 'tarefa']].head(15)
```
